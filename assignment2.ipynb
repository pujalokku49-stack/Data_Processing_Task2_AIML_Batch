```python
# creating numpy array and performing it's operations

import numpy as np

math_scores = df['math_score'].values
math_array = np.array(math_scores)
```

```python
mean = np.mean(math_array)
median = np.median(math_array)
maximum = np.max(math_array)
minimum = np.min(math_array)

print(mean, median, maximum, minimum)
```

```python
normalized_scores = (math_array - minimum) / (maximum - minimum)
print(normalized_scores)
```

```python
# pandas exploration

import pandas as pd

df = pd.read_csv("student_scores.csv")
```

```python
low_attendance = df[df['attendance'] < 70]
print(low_attendance)
```

```python
# data preprocessing
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df['math_score'] = df['math_score'].astype(float)
```

```python
df['exam_date'] = pd.to_datetime(df['exam_date'], errors='coerce')
```

```python
Q1 = df['math_score'].quantile(0.25)
Q3 = df['math_score'].quantile(0.75)
IQR = Q3 - Q1

df = df[(df['math_score'] >= Q1 - 1.5*IQR) & 
        (df['math_score'] <= Q3 + 1.5*IQR)]
```

```python
df.drop_duplicates(inplace=True)
```

```python
# data analysis
df['average_score'] = (df['math_score'] + df['science_score']) / 2
top5 = df.sort_values(by='average_score', ascending=False).head(5)
print(top5)
```

```python
correlation = df['attendance'].corr(df['average_score'])
print(correlation)
```

```python
grouped = df.groupby('gender')['average_score'].mean()
print(grouped)
```
